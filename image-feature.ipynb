```python
import os
import numpy as np

from pickle import dump
from keras import backend as K
from keras.models import Model
from keras.applications.vgg16 import VGG16, preprocess_input
from tensorflow.keras.preprocessing.image import load_img, img_to_array
```

```python
ROOT_PATH = "drive/MyDrive/MLproject/Automatic-emotional-image-captioning-main/flickr5k"
IMAGES_PATH = "drive/MyDrive/MLproject/Automatic-emotional-image-captioning-main/flickr5k/images"
```

```python
# Load and restructure vgg16 model
model = VGG16()
model = Model(inputs=model.inputs, outputs=model.layers[-2].output)

image_features = dict()
for img_name in os.listdir(IMAGES_PATH):
    image = load_img(os.path.join(IMAGES_PATH, img_name),target_size=(224, 224))

    # Change to array, expand a new dim, preprocess_input and extract feature
    image = img_to_array(image)
    image = np.expand_dims(image, axis=0)
    image = preprocess_input(image)
    feature = model.predict(image, verbose=0)

    image_features[img_name] = feature

# Store image_features
dump(image_features, open(os.path.join(ROOT_PATH,'flickr5k_features.pkl'),'wb'))
```
