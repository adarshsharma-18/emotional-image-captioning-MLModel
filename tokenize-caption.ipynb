```python
import os
from pickle import dump
from nltk import flatten
# Import Tokenizer from tensorflow.keras.preprocessing.text instead of keras.preprocessing.text
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
ROOT_PATH = "/drive/MyDrive/MLproject/Automatic-emotional-image-captioning-main/flickr5k"
CAPTIONS_PATH = "drive/MyDrive/MLproject/Automatic-emotional-image-captioning-main/flickr5k/captions.txt"
```

```python
captions_dic = dict()

f = open(CAPTIONS_PATH, 'r')
for line in f:
    img_name = line.split(',')[0]
    caption = line.split(',')[1].rstrip('\n')
    caption = caption.replace(' .','').lower()
    caption = 'captionbegin ' + caption + ' captionend'

    if img_name not in captions_dic:
        captions_dic[img_name] = []

    captions_dic[img_name].append(caption)
```

```python
# Create and fit tokenizer
tokenizer = Tokenizer()

all_captions = flatten(list(captions_dic.values()))
tokenizer.fit_on_texts(all_captions)
# Ensure the directory exists before attempting to write the file
os.makedirs(ROOT_PATH, exist_ok=True) # Create directory if it doesn't exist

# Store tokenizer
dump(tokenizer, open(os.path.join(ROOT_PATH,'flickr5k_tokenizer.pkl'),'wb'))
```

```python
vocab_length = len(tokenizer.word_index) + 1
vocab_length
```

```python
max_words_length = 0
for caption in all_captions:
    if len(caption.split()) > max_words_length:
        max_words_length = len(caption.split())
max_words_length
```
